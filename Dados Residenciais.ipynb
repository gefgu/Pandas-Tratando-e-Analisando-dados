## Dados Residenciais

```python
import pandas as pd
```

```python
dados = pd.read_csv("dados/aluguel.csv", sep=";")
dados.head()
```

```python
list(dados["Tipo"].drop_duplicates())
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa de Vila',
 'Chácara']
residencial
```

```python
selecao_residencial = dados["Tipo"].isin(residencial)
```

```python
dados_residenciais = dados[selecao_residencial]
dados_residenciais
```

```python
dados.shape
```

```python
dados_residenciais.shape
```

```python
dados_residenciais.index = range(dados_residenciais.shape[0])
```
