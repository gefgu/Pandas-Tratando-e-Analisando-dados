## Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv("dados/residencias.csv", sep=';')
dados.head(10)
```

```python
dados["Valor Bruto"] = dados["Valor"] + dados["Condominio"] + dados["IPTU"]
dados.head(15)
```

```python
dados["Valor por m^2"] = (dados["Valor"] / dados["Area"]).round(2)
dados.head(15)
```

```python
dados["Valor Bruto por m^2"] = (dados["Valor Bruto"] / dados["Area"]).round(2)
dados.head(15)
```

```python
casa = ["Casa", "Casa de Condomínio", "Casa de Vila"]
```

```python
dados["Tipo Agregado"] = dados["Tipo"].apply(lambda x: "Casa" if x in casa else "Apartamento")
dados.head(15)
```

## Excluíndo valores

```python
df_aux = dados[["Valor Bruto", "Tipo Agregado", "Valor por m^2", "Valor Bruto por m^2"]]
df_aux.head(10)
```

```python
del df_aux["Valor Bruto"]
df_aux.head(10)
```

```python
df_aux.pop("Valor Bruto por m^2")
```

```python
df_aux.head()
```

```python
dados.drop(["Valor Bruto por m^2", "Valor Bruto"], axis=1, inplace=True)
```
